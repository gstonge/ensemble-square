```python
# calculate MAPE

startdates = [ Week(2021,18), Week(2021,22), Week(2021,27), Week(2021,37), Week(2021,51),
             Week(2022,2), Week(2022,11), Week(2022,23), Week(2022,31), Week(2022,44)]


enddates = [ Week(2021,43), Week(2021,47), Week(2021,52), Week(2022,9),  Week(2022, 10),
           Week(2022, 13), Week(2023, 11), Week(2023, 23), Week(2023, 18), Week(2023, 17)]

predsall = [ predictions_rd5, predictions_rd6, predictions_rd7, predictions_rd9, 
            predictions_rd11, predictions_rd12, predictions_rd13, predictions_rd14, predictions_rd15, predictions_rd16]




rds = [ 5, 6, 7, 9,  11, 12, 13, 14 ,15, 16]

#rds = [14]

ind = 0


for predrd in predsall:
    rd = rds[ind]
    
    start_week = startdates[ind]
    end_week = enddates[ind]
    
    ind += 1
    
    if rd in [4, 5]:
        max_date = datetime.datetime(2021, 6, 19)
    elif rd in [6,7,8,9]:
        max_date = datetime.datetime(2021, 12, 18)
    else:
        max_date = None


    for target in ['case', 'death', 'hosp']:
    #for target in ['case']:

        dfpt = pd.read_pickle(f'allensemble_mape_{target}_rd{rd}_517_all.pkl')


        obsdf = pd.DataFrame()


        for l in loclist:    
            target_location = l

            if target == 'hosp':
                target_obs = 'hospitalization'
            else:
                target_obs = target
            observations = pd.read_parquet(f"./dat/truth_{'inc' if incidence else 'cum'}_{target_obs}.pq")
            observations['date'] = pd.to_datetime(observations['date'])

            #filter start - end week
            observations = observations[(observations['date'] >= pd.to_datetime(start_week.startdate())) & 
                                    (observations['date'] <= pd.to_datetime(end_week.enddate()))]



            #filter location
            observations = observations[observations['location'] == target_location]

            #aggregate to weekly
            observations = observations.groupby(['location', pd.Grouper(key='date', freq='W-SAT')]).sum().reset_index()

            if max_date:
                observations = observations[observations['date'] <= max_date].copy()

            #transform to Observation object
            observations = Observations(observations)


            obsdf = pd.concat([obsdf, observations])

        obsdf = obsdf.reset_index()
        obsdf = obsdf.astype({'location':'int'})
        obsdf = obsdf.merge(locations, how='inner', on='location')
        obsdf = obsdf.drop(['index'], axis=1)


        obs = obsdf

        models = list(dfpt.Model.unique())
        
        locs = list(dfpt.Location.unique())

        mapedf = pd.DataFrame(columns=['Model', 'Location','Population', 'MAPE'])

        for model in models:
            for l in locs:
                a = dfpt[(dfpt.Model ==model)& (dfpt.Location ==l)]
                b = obs[obs.location_name ==l]

                n = b.shape[0]

                mapesum = 0
                realvals = list(b.value)
                predvals = list(a.value)

                if len(predvals) == 0:
                    continue
                    
                for i in range(n):
                    if realvals[i] == 0:
                        n = n - 1
                        continue
                    
                    
                    err = abs((realvals[i]-predvals[i])/realvals[i])

                    mapesum += err
                    
                if n == 0:
                    mape = None
                else:
                    mape = 1/n * mapesum


                data = {'Model': model,'Location': l,'Population': a.population.unique()[0], 'MAPE':mape}

                # Create DataFrame
                newdf = pd.DataFrame(data, index=[1])

                mapedf = pd.concat([mapedf, newdf])

        mapedf = mapedf.reset_index()
        mapedf = mapedf.drop(['index'], axis=1)
        
        if target == 'hosp':
            mapedf_hosp = mapedf.copy()
        elif target =='death':
            mapedf_death = mapedf.copy()
        elif target == 'case':
            mapedf_case = mapedf.copy()
        
        
    mapedf_hosp.to_pickle(f'mapeval_hosp_rd{rd}_517_all.pkl')
    mapedf_death.to_pickle(f'mapeval_death_rd{rd}_517_all.pkl')
    mapedf_case.to_pickle(f'mapeval_case_rd{rd}_517_all.pkl')
        
        
```

```python
# WIS rescaled by standard deviation at each week 
rds = [5,6,7,9,11,12,13,14,15,16]

wisstdev_case = pd.DataFrame()
wisstdev_death = pd.DataFrame()
wisstdev_hosp = pd.DataFrame()

for rd in rds:
    print(rd)
    for target in ["case", 'death', "hosp"]:
        df = pd.read_pickle(f'allmodel_wis_{target}_rd{rd}_530_weeks.pkl')  
        for loc in list(df.Location.unique()):
            for date in list(df.Date.unique()):
                df1 = df[df.Location == loc]
                df1 = df1[df1.Date == date]
                
                stdev = df1['WIS'].std()
                
                df1['WISscale'] = df1['WIS'] / stdev
                
                if target == 'case':
                    wisstdev_case = pd.concat([wisstdev_case, df1])
                elif target == 'death':
                    wisstdev_death = pd.concat([wisstdev_death, df1])
                elif target =='hosp':
                    wisstdev_hosp = pd.concat([wisstdev_hosp, df1])
wisstdev_case.reset_index() 
wisstdev_death.reset_index() 
wisstdev_hosp.reset_index() 

wisstdev_case.to_pickle(f'wis_stdev_wks_case_531.pkl') 
wisstdev_death.to_pickle(f'wis_stdev_wks_death_531.pkl') 
wisstdev_hosp.to_pickle(f'wis_stdev_wks_hosp_531.pkl') 
```
